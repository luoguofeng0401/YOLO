```python
import cv2
print(cv2.__version__)               #查看版本

def extractImages(pathIn, pathOut):  #提取圖像函數
    count = 0
    vidcap = cv2.VideoCapture(pathIn)#建立一個 VideoCapture 物件,這個物件會連接到視頻
    success,image = vidcap.read()
    #使用 read 函數來擷取一張張連續的畫面影像(第一個傳回值代表成功與否,第二個傳回值 frame 就是攝影機的單張畫面)
    success = True
    while success:
        vidcap.set(cv2.CAP_PROP_POS_MSEC,(count*200))    # added this line  (count*500(毫秒msec))
        success,image = vidcap.read()                    
        #使用 read 函數來擷取一張張連續的畫面影像(第一個傳回值代表成功與否,第二個傳回值 frame 就是攝影機的單張畫面)
        print ('Read a new C1P102_2018_07_25_coding1_frame%d:'% count, success)            #印出成功與否
        cv2.imwrite( pathOut + "\\C1P102_2018_07_25_coding1_frame%d.jpg" % count, image)  # save frame as JPEG file
                                    #攝影機編號_製作日期_編碼_流水號
        count = count + 1     
extractImages('test.mp4', 'Images')
```
